## Big Normal Examples

```python
ohlc_data_day1=np.array([
    [100.0,101.5,99.5,101.0],
    [101.0,102.0,100.5,101.8],
    [101.8,103,101.0,102.5]
])
ohlc_data_day2=np.array([
    [200.0,201.5,299.5,201.0],
    [201.0,202.0,200.5,201.8],
    [201.8,203,201.0,202.5],
    [201.8,203,201.0,202.5]
])
```

```python
import numpy as np
```

```python
np.max(ohlc_data_day1)
np.min(ohlc_data_day1)
np.sum(ohlc_data_day1)
```

```python
# scaler operations
ohlc_data_day1*2
```

```python
#relational
ohlc_data_day1 == 100
```

```python
# vector operations arthimetic
ohlc_data_day1 * ohlc_data_day2
```

```python
#mean/median/std/var
np.mean(ohlc_data_day1)
np.median(ohlc_data_day1)
np.std(ohlc_data_day1)
np.var(ohlc_data_day1)
```

```python
#Trignometric functions
np.cos(ohlc_data_day1)
```

```python
#dot_product
# col of 1st metric should be equal to rows second metric
np.dot(ohlc_data_day1,ohlc_data_day2)
```

```python
#log and exponents
np.exp(ohlc_data_day1)
```

```python
#round/floor/ceil
np.round(ohlc_data_day1)
np.floor(ohlc_data_day1)
np.ceil(ohlc_data_day1)
```
