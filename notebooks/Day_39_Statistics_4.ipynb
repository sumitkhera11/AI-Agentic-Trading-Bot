```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Fetch historical data
tickers = ['MSFT', 'AAPL']
data = yf.download(tickers, start='2023-01-01', end='2023-06-30',auto_adjust=False)['Adj Close']

# Calculate daily returns
returns = data.pct_change().dropna()

# Calculate covariance
cov_matrix = returns.cov()

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(returns['MSFT'], returns['AAPL'], alpha=0.5)
plt.title('Scatter Plot of Daily Returns: MSFT vs AAPL')
plt.xlabel('MSFT Returns')
plt.ylabel('AAPL Returns')
plt.grid(True)

plt.tight_layout()
plt.show()

print("Covariance Matrix:\n", cov_matrix)
```

```python
# above o/p is +ve
```
