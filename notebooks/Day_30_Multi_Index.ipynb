```python
m_index = [("AAPL",1),("MSFT",2),("META",3),("TCS",4),("IOCL",5)]
m_ser = pd.Series([100,200,300,400,500],index=m_index)
m_df = pd.DataFrame([100,200,300,400,500],index=m_index)
print(m_ser)
print(m_df)
```

```python
# it will not work in case of df
# m_df[('AAPL', 1)]


# but in case of df - use this technique
# iloc position based indexing use karta hai.
# [row_position , column_position]
m_df.iloc[0,0]
m_df.iloc[1,0]


# Agar actual index value se access karna ho:
m_df.loc[[('MSFT',2)]]
m_df.loc[[('MSFT',2)]].iloc[0,0]

# iloc → position based
# loc → actual label/index based
```

## actual code of multiindex

```python
# Multiindex

# Multiindex from tuple
t_index = pd.MultiIndex.from_tuples(m_df.index)
t_index

# m_df.loc[('MSFT',2)]


# Multiindex from product
pp_index = [['AAPL','MSFT','META'],[100,200]]
p_index = pd.MultiIndex.from_product(pp_index)
print(p_index)
```

```python
t_index
```

```python
# check levels in Multiindex
# m_df.index.level[0]
t_index.levels 
t_index.levels[0]
t_index.levels[1]
```

```python
p_index.levels
p_index.levels[0]
p_index.levels[1]
```
