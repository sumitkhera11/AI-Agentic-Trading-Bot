```python
# melt -> real world example

import pandas as pd
import yfinance as yf

data =  yf.download(tickers=['AAPL','AMZN','MSFT','META'], start= '2024-01-01', end= '2024-06-30',group_by= 'Ticker',auto_adjust=False)
# Future_stack = True -> it removes warning 
# .reset_index()  - by default level 0 index is set - it gives default index
#  but need of level 1 index
data = data.stack(level=0, future_stack=True).reset_index(level=1)
data = data.reset_index()
# data = data.set_index('Date')
melted_df =data.melt(id_vars=['Date','Ticker'], var_name = 'Metrics', value_name = 'Value',value_vars=['Open',	'High',	'Low','Close','Adj Close','Volume'])
melted_df
 
```

```python
data.groupby('Ticker')['Open'].mean()

# return mean of every day of each ticker
data.groupby(['Ticker','Date'])['Open'].mean()

#unstack()  = rows becomes cols  - the same thing can acheive with the pivot table
data.groupby(['Ticker','Date'])['Open'].mean().unstack()

# or
data.pivot_table(index='Ticker',columns='Date', values='Open')
# groupby or pivot_table both gives same result but pivot_table is more efficient
```

```python
# by default it calulates the mean value
# it works on single index

# Pivot table = Rows aur columns ko rearrange karke
# summary table banana
data.pivot_table(index='Ticker',columns='Date', values='Open')
# or
# by default it cal on mean value
# pivot_table applies on - 1 col only i.e  values='Open'
data.pivot_table(index='Ticker',columns='Date', values='Open', aggfunc='mean')

# in short - pivot_table is the advance of groupby

# it applies on all the columns
data.pivot_table(index='Ticker',columns='Date', aggfunc='mean')

# can diff aggregate function by mentioning on column name
data.pivot_table(index='Ticker',columns='Date', aggfunc={'Open':'mean','High':'max','Close':'min'})


# ------------
# it works on multi index - Ticker Date
data.pivot_table(index=['Ticker','Date'],columns='Date', aggfunc={'Open':'mean','High':'max','Close':'min'})
```

```python
# margins
data.pivot_table(index='Ticker',columns='Volume', aggfunc='sum',margins=True)
```

```python
# on y axis prices 

# Data reshape kar raha hai (pivot_table)
# Close prices organize kar raha hai
# Graph bana raha hai (plot())

# Pivot table = Rows aur columns ko rearrange karke
# summary table banana

# index='Date'  - means-> Date ko row labels bana do

# columns='Ticker'  - Ticker names ko columns bana do
#             AAPL    MSFT    META    AMZN
# Date

# values='Close'  - Sirf Close column ki values use karo
# Date         AAPL    MSFT    META    AMZN
# 2024-01-02   185.6   363.8   343.2   149.9
# 2024-01-03   186.1   364.2   345.1   151.0
# 2024-01-04   184.3   361.5   341.7   152.3

# data.pivot_table(index='Date',columns='Ticker',values='Close').plot()

import matplotlib.pyplot as plt

chart = data.pivot_table(index='Date',columns='Ticker',values='Close')
chart.plot()

plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.title("Stock Price Trend")

plt.show()
```
