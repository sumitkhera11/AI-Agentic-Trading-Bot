## Normal Distribution

```python
import pandas as pd
import numpy as np
from scipy.stats import norm
```

#### Find Stand Deviation

```python
salaries = [5,5,5,6,4,4,5,5,5,5,5,5,6,5]
sal_ser = pd.Series(salaries)
sal_ser
```

```python
# Mean
sal_ser.mean()
```

```python
# Standard Deviation
sal_ser.std()
```

### Find Cumulative Probability


```python
# first parameter  x = 0
# Second parameter mean = 0
# third parameter s.d   = 1
norm.cdf(0,0,1)
```

```python
# first parameter  x = -1
# Second parameter mean = 0
# third parameter s.d   = 1
norm.cdf(-1,0,1)
```
