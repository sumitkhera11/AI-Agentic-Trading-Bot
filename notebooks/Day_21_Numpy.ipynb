## Advance Indexing and Broadcasting

```python
ohlc1=np.array([
    [100.0,101.5,99.5,101.0],
    [101.0,102.0,100.5,101.8],
    [101.8,103,101.0,102.5]
])
ohlc2=np.array([
    [200.0,201.5,299.5,201.0],
    [201.0,202.0,200.5,201.8],
    [201.8,203,201.0,202.5],
    [201.8,203,201.0,202.5]
])
```

```python
ohlc1[1,2]
ohlc1[1:,2:]
ohlc1[[1],[2]]
ohlc1[1:,[2,3]]
```

## Fancy Indexing

```python
ohlc2[[0,3],1:3]
```

## Boolean Fancy

```python
ohlc1>102
ohlc1[ohlc1>102]
```

```python
ohlc2>102
ohlc2[ohlc2>202]
```

```python
ohlc1 % 2==0
ohlc1[ohlc1 % 2==0]
```

```python
ohlc2[(ohlc2 % 2==0.0) & (ohlc2> 201)]
```

## Boradcasting

```python
ohlc3=np.array([
    [100.0,101.5,99.5,101.0],
    [101.0,102.0,100.5,101.8],
    [101.8,103,101.0,102.5]
])
ohlc4=np.array([
    [200.0,201.5,299.5,201.0]
])
```

```python
ohlc3+ohlc4
```

```python
#mean squared error
def msc(ohlc3,ohlc4):
    return np.mean((ohlc3-ohlc4)**2)
msc(ohlc3,ohlc4)
```
