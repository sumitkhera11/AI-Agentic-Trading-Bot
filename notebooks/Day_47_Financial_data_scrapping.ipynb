```python
def CryptoPrices(symbol):
    url = f"https://sg.finance.yahoo.com/quote/{symbol}-USD"
    headers = {
        "User-Agent":"Mozilla/5.0"
    }
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text,'html.parser')
    price = soup.find('span', {'data-testid':'qsp-price'}).text
    change = soup.find('span', {'data-testid':'qsp-price-change'}).text
    per_change = soup.find('span', {'data-testid':'qsp-price-change-percent'}).text
    details = {'Symbol': symbol, 'Price': price, 'Change': change, 'Percentage Change' : per_change}
    return details

```

```python
def CryptoPrices_New(symbol):

    url = f"https://sg.finance.yahoo.com/quote/{symbol}-USD"

    headers = {
        "User-Agent":"Mozilla/5.0"
    }

    response = requests.get(url, headers=headers)

    soup = BeautifulSoup(response.text,'html.parser')

    price_tag = soup.find('span',{'data-testid':'qsp-price'})

    if price_tag:
        price = price_tag.text
        print(symbol, "->", price)

    else:
        print(symbol, "FAILED")
        print(response.status_code)

        return None

    return price
```

```python
CryptoPrices_New('ETH')
CryptoPrices_New('SOL')
```

```python
MyCrypto = ['BTC','ETH','XRP','BNB','SOL']
CryptoData = []
```

```python
for symbol in MyCrypto:
    CryptoData.append(CryptoPrices_New(symbol))
    print(f"Getting Price of Awesome Crypto: {symbol}")
CryptoData
```

```python
def FetchPrices():
    MyCrypto = ['BTC','ETH','XRP','BNB','SOL']
    CryptoData = []
    for symbol in MyCrypto:
        CryptoData.append(CryptoPrices_New(symbol))
        print(f"Getting Price of Awesome Crypto: {symbol}")
    return CryptoData
```

```python
while True:
    FetchPrices()
    time.sleep(5)
```
